```python
import requests
import os
from bs4 import BeautifulSoup
from selenium import webdriver

browser = webdriver.Firefox()
browser.get("http://mathege.ru/")
browser.add_cookie(
    {'path': '/or/ege/', 'secure': False, 'domain': 'mathege.ru', 'value': '2',
     'httpOnly': False, 'name': 'cookieEgeLevel', 'expiry': None})

s = requests.Session()
cookies = {'cookieEgeLevel' : '2'}
r = s.get('http://mathege.ru/or/ege/ShowProblems?offsetStr=577', cookies=cookies)
soup = BeautifulSoup(r.content, 'html.parser')
number_pages_text = soup.findAll(text=re.compile("Просмотр выбранных заданий с"))
start_task, last_task = list(map(int, re.findall('\d+', number_pages_text[0])))
last_task = 12 + start_task
while start_task < last_task:
    r = s.get(
        'http://mathege.ru/or/ege/ShowProblems?offsetStr={}'.format(start_task),
        cookies=cookies)
    soup = BeautifulSoup(r.content, 'html.parser')


#print(soup.findAll(id="QuizBody"))
    quizbody = soup.find(id="QuizBody")
    task_number = 0
    while quizbody:
#print(td)
        table = quizbody.findParent('table')
        prototype = table.find(text=re.compile("Прототип")).findNext('td').find('a').text
        type_number_task_text = table.findPrevious('table').find('h2').text
        print_href = 'http://mathege.ru{}'.format(table.findPrevious('table').find('h2').findParent('td').findNext('td').a['href'])
        #print(print_href)
        browser.get(print_href)
        browser.save_screenshot(os.path.join(os.getcwd(), type_task, prototype,'{}.png'.format(number_task)))
        
        #print(type_number_task_text)
        type_task, number_task = re.findall('\d+', type_number_task_text)

        #print(task_number, start_task, type_task, number_task, prototype)
        #print(quizbody.findAll("img"))
        
        quizbody = quizbody.findNext(id="QuizBody")
        task_number += 1
        if not os.path.exists(type_task):
            os.makedirs(type_task)
        if not os.path.exists(os.path.join(type_task, prototype)):
            os.makedirs(os.path.join(type_task, prototype))
        with open(os.path.join(type_task, prototype, number_task), 'w') as file:
            file.write(table.prettify())

    start_task += 5
```
